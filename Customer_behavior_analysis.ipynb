The  review rating has null values
- we an fill it with the mean or median values.
- coming  to the mean it has outliers so the distribution will have skew
- filling it with median is possible, 
    - but the data has different categories like shoes, accesories, clothing.
    - so if we take the median value for all the null values,
    - the category wise rating will affect the overall purchasing products.
    - so take the median for catergorywise.

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

# this is in text form so we need to make it in numerical so it is easy to analyze the data

```python
frequency_mapping = {
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
